```python
import json


print('''
==============================================WELCOME TO DICTIONARY ===============================================
        ''')
ch=1

while(ch<=4):
    print('''
        Enter the choice 
            1. To Add a new word
            2. To Find the meaning
            3. To Update a word
            4. To Exit''')
    print()
    ch=int(input ("Enter the choice : "))
    print()
    if ch==1:
        f= open('words.txt','r')
        d=json.load(f)
        f.close()
        print("---------- ADDING--------- ")
        #print(d)
        print()
        len1=int(input("How many entries you want add: "))
        print()
        print("Enter the word and meaning")
        for i in range (len1):
            d.update({input():input()})
        f= open('words.txt','w')    
        json.dump(d,f)
        f.close()
        print("ADDED LIST IS ",d)    
    elif ch==2:
        print("---------- SEARCHING--------- ")
        f= open('words.txt','r')
        d=json.load(f)
        #print(d)
        print()
        dn=input("Enter the word to display meaning: ")
        if dn in d.keys():
            print()
            print("The word meaning for ",dn, ' is : ',d[dn])  
        else:
            print()
            print(dn , "not found in list") 
        f.close()        
    elif ch==3:
        print("---------- EDITING--------- ")
        #print(d)
        print()
        n=input("Enter a word to edit : ")
        print()
        f= open('words.txt','r')
        d=json.load(f)
        f.close()
        if n in d.keys():
            meaning=input("Enter the meaning : ")
            d.update({n:meaning})
            f= open('words.txt','w')    
            json.dump(d,f)
            f.close()
            print("EDITED DICTIONARY IS \n",d)
        else:   
            print(n , "not found in list")
    elif ch==4:
        print()
        print("You have opted to Exit......\n")
        print('#'*25,"Thank you",'#'*25)
        break
    else:
        print()
        print("Please enter a correct option ")
        ch=4 

            
```
